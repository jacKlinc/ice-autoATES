```python
validation_dir = Path.cwd().parent / "data" / "validation"
```

## Mount Seymour

```python
mount_seymour_kmz = validation_dir / "mount-seymour" / "layers.kmz"
v = Validator(mount_seymour_kmz, simple).run()
```

```python
evaluate.report(v.predicted, v.truth, save_dir=mount_seymour_kmz.parent)
```

```python
evaluate.plot_confusion_matrix(v.predicted, v.truth)
```

```python
evaluate.plot_side_by_side(v.predicted, v.truth, predicted_title="Simple model")
```
